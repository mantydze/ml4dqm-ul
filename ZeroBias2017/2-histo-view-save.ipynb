```python
import os 
import pandas as pd
from utils import view_histo_from_raw
```

```python
hname = "chargeInner_PXLayer_1"
run_number = 299326
```

```python
source_path = "/home/mantydze/data/ZeroBias2017B/{hname}.csv".format(hname=hname)
target_path = "/home/mantydze/data/ZeroBias2017B/{hname}/{run_number}/".format(hname=hname, run_number=run_number)

df = pd.read_csv(source_path)

# Create Run directory if rows exists
if len(df[df["run"] == run_number]) > 0:
    
    
    if not os.path.exists(target_path):
        os.makedirs(target_path) 

for index, row in df[df["run"] == run_number].iterrows():
    filename = "{run}_{lumi}_{good}.png".format(target_path=target_path, 
                                               run=row["run"],
                                               lumi=row["lumi"], 
                                               good=row["good"])
    
    save_path = os.path.join(target_path, filename)
    
    view_histo_from_raw(row, save_path=save_path)
```
