```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
training_dir = "/home/mantydze/mokslai/ml4dqm-ul/ZeroBias2017B/scripts/trainings/data_ext/t-0.9-all"
df_train_path = Path(training_dir, 'df_train.csv') # classified dataset
df_anom_path = Path(training_dir, 'df_anomalies.csv')
df_orig_path = "/home/mantydze/mokslai/ml4dqm-ul/ZeroBias2017B/scripts/data_raw/chargeInner_PXLayer_1.csv"
```

```python
df_train = pd.read_csv(df_train_path)
df_anom = pd.read_csv(df_anom_path)
df_orig = pd.read_csv(df_orig_path)
```

```python
df_orig.drop(df_orig[df_orig.entries == 0].index, inplace=True)

nbins = 100
step = 1

bin_cols = [f"bin_{i}" for i in range(1, nbins+1, step)]
bin_cols = bin_cols[1:-1]

# X = df.filter(bin_cols, axis=1).copy().div(df.entries, axis=0)
# y = df["y"]  
```

```python
def plot_raw(df, color="orange", title=""):
    plt.figure(figsize=(20,10))
    plt.title(title)
    for _i, row in df.iterrows():
        data = [row[col] for col in bin_cols]
        plt.plot(range(len(data)), data, color=color, alpha=0.1)
    
def plot(df, color, title=""):
    plt.figure(figsize=(20,10))
    plt.title(title)
    for _i, row in df.iterrows():
        data = [row[col]/row["entries"] for col in bin_cols]
    
#         plt.plot(range(len(data)), data, color=palette[row['y']])
        plt.plot(range(len(data)), data, color=color, alpha=0.05)
        
```

```python
# GOOD PIXEL
plot(df_orig[df_orig['good_pixel']==1], 'green', "GOOD PIXEL")
```

```python
# BAD PIXEL
plot(df_orig[df_orig['good_pixel']==0], 'red', "BAD PIXEL")
```

```python
# Golden JSON GOOD
plot(df_orig[df_orig['good']==1], 'green', "GOLDEN JSON GOOD")
```

```python
# Golden JSON BAD
plot(df_orig[df_orig['good']==0], 'red', "GOLDEN JSON BAD")
```

```python
# BAD pixel match predicted
bmatch = df_train[(df_train["good_pixel"]==0) & (df_train["y"]==0)]
plot(bmatch, 'red', "BAD PIXEL - predicted BAD")
```

```python
# BAD pixel not match predicted
bmatch = df_train[(df_train["good_pixel"]==0) & (df_train["y"]==1)]
plot(bmatch, 'red', "BAD PIXEL - predicted GOOD")
```

```python
gmatch = df_train[(df_train["good_pixel"]==1) & (df_train["y"]==1)]
plot(gmatch, 'red', "GOOD PIXEL - predicted GOOD")
```

```python
nmatch = df_train[df_train["good_pixel"]!=df_train["y"]]
plot(nmatch, 'red', "PIXEL != predicted")
```
